```python
#dropped since they just had one value
df=df.drop('delinq_amnt',axis=1)
df=df.drop('chargeoff_within_12_mths',axis=1)
df=df.drop('acc_now_delinq',axis=1)
df=df.drop('collections_12_mths_ex_med',axis=1)
df=df.drop('tax_liens',axis=1)
df=df.drop('policy_code',axis=1)
df=df.drop('application_type',axis=1)
df=df.drop('initial_list_status',axis=1)
df=df.drop('pymnt_plan',axis=1)
```

```python
contvarlist=[]
chkcontvarlist=[]
for columns in df:
    if df[columns].nunique()>10 and (df[columns].dtype=='int64' or df[columns].dtype=='float64'):
        contvarlist.append(columns)
    elif df[columns].dtype=='int64' or df[columns].dtype=='float64':
        chkcontvarlist.append(columns)
    
print(contvarlist)
print('\n'+'_______________________________________________________________')
print(chkcontvarlist)
```

```python
catvarlist=[]
chkcatvarlist=[]
for columns in df:
    if df[columns].nunique()<20 and (df[columns].dtype=='object'):
        catvarlist.append(columns)
    elif df[columns].dtype=='object':
        chkcatvarlist.append(columns)

    
print(catvarlist)
print('\n'+'_______________________________________________________________')
print(chkcatvarlist)
```
